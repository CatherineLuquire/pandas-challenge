
## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#format dataframe
pd.options.display.float_format = '${:,.2f}'.format

# number of purchases by gender 
gender_counts = purchase_data["Gender"].value_counts()

#create dataframe with relevant gender and purchase data
clean_gender_df = purchase_data[["Gender", "Price"]]

# set gender to index
gender_pa_df = clean_gender_df.set_index("Gender")

#set df variables 
male_purchases = gender_pa_df.loc["Male", "Price"]
total_male_purchases = male_purchases.sum()
avg_val_male_purchases = male_purchases.mean()
avg_purchase_per_male = total_male_purchases/gender_demo["Male"]

female_purchases = gender_pa_df.loc["Female", "Price"]
total_female_purchases = female_purchases.sum()
avg_val_female_purchases = female_purchases.mean()
avg_purchase_per_female = total_female_purchases/gender_demo["Female"]

nd_purchases = gender_pa_df.loc["Other / Non-Disclosed", "Price"]
total_nd_purchases = nd_purchases.sum()
avg_val_nd_purchases = nd_purchases.mean()
avg_purchase_per_nd = total_nd_purchases/gender_demo["Other / Non-Disclosed"]

# build dataframe
gender_df = pd.DataFrame({"Gender": ["Female", "Male", "Other / Non-Disclosed"],
                          "Purchase Count": [gender_counts["Female"], gender_counts["Male"], gender_counts["Other / Non-Disclosed"]],
                          "Average Purchase Price": [avg_val_female_purchases, avg_val_male_purchases, avg_val_nd_purchases],
                          "Total Purchase Value": [total_female_purchases, total_male_purchases, total_nd_purchases],
                          "Average Purchase per Person": [avg_purchase_per_female, avg_purchase_per_male, avg_purchase_per_nd]})

gender_df_final =gender_df.set_index("Gender")

gender_df_final
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# format dataframe
pd.options.display.float_format = '{:,.2f}%'.format

# create df for relevant data
age_demo = purchase_data[["Age", "SN"]]

# drop dupliate data
age_demo_clean = age_demo.drop_duplicates("SN")

# assign bins 
age_clean = age_demo_clean[["Age"]]
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_clean["Bins"] = pd.cut(age_demo["Age"], bins, labels=group_names, include_lowest=True)

# set index to bins
age_demo_index = age_clean.set_index("Bins")
age_demo_index

# count ages in bins 
age_demo_counts = age_demo_index.groupby("Bins").count()
age_demo_counts

# build dataframe 
age_pa_df = pd.DataFrame({"": group_names,
                            "Total Count": (age_demo_counts["Age"]), 
                            "Percentage of Players": (age_demo_counts["Age"]/unique_sn)*100})

# set index to bins
age_pa_df_final = age_pa_df.set_index("")
age_pa_df_final
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# format dataframe
pd.options.display.float_format = '${:,.2f}'.format

# clean dataframe for relevant data
age_pa = purchase_data[["Age", "Price"]]

# sort data into bins
age_pa["Bins"] = pd.cut(age_pa["Age"], bins, labels=group_names, include_lowest=True)

# set index to bins
age_pa = age_pa.set_index("Bins")

# groupby bins to count data by age
age_pa_counts = age_pa.groupby("Bins").count()

# groupby bins to average price by age
age_pa_avg = age_pa.groupby("Bins").mean()

# groupby bins to find total value spent price by age
age_pa_sum = age_pa.groupby("Bins").sum()

# build dataframe
age_pa_df = pd.DataFrame({"Age Ranges": group_names, 
                          "Purchase Count": (age_pa_counts["Age"]), 
                          "Average Purchase Price": age_pa_avg["Price"],
                          "Total Purchase Value": age_pa_sum["Price"],
                          "Average Total Purchase per Person": (age_pa_sum["Price"]/(age_demo_counts["Age"]))})

# set index to age ranges
age_pa_df = age_pa_df.set_index("Age Ranges")
age_pa_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# build dataframe with relevant data
spenders_df = purchase_data[["SN", "Price"]]
spenders_df

#to get top spender info
spenders_combined_df = spenders_df.groupby(["SN"]).sum()["Price"]

# to get spender purchase count
spenders_count = spenders_df.groupby(["SN"]).count()["Price"]

# build spenders dataframe
spenders_df_values = pd.DataFrame({"Purchase Count": spenders_count, 
                                  "Average Purchase Price": (spenders_combined_df/spenders_count),
                                  "Total Purchase Value": spenders_combined_df})

# sort dataframe by max purchase value for top spenders
spenders_final = spenders_df_values.sort_values(by='Total Purchase Value', ascending=False)
spenders_final.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# create dataframe with relevant data
pop_items = purchase_data[["Item ID", "Item Name", "Price"]]
pop_items

# groupby item name and ID
pop_items_group = pop_items.groupby(["Item ID", "Item Name"])
pop_items_group.head()

#find total item count
item_count = pop_items_group["Item Name"].count()
item_count

# find total $ spend on items
total_values = pop_items_group["Price"].sum()
total_values

# find average purchase price per item
item_price = pop_items_group["Price"].mean()
item_price

# build dataframe
pop_items_df = pd.DataFrame({"Purchase Count": item_count,
                             "Item Price": item_price,
                             "Total Purchase Value": total_values})

# sort by the number of time item purchased
pop_items_final = pop_items_df.sort_values(by='Purchase Count', ascending=False)
pop_items_final.head()
```
